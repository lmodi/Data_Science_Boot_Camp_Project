```python
import os
import json
import pandas as pd

# Data is in the data directory in project_1 folder
DATA_DIR_MOV = os.path.join('data', 'boxofficemojo')

# 'listdir' function from 'os'  gives you a list of all
target_file_names_Mov = os.listdir(DATA_DIR_MOV)

movie = list()
for files in target_file_names_Mov:
    target_file_path = os.path.join(DATA_DIR_MOV, files)
    with open(target_file_path,'r') as target_file:
        movie.append(json.load(target_file))
        

movies_df = pd.DataFrame(movie)        
```

```python
len(movies_df)
```

```python
DATA_DIR_CRI = os.path.join('data', 'metacritic')

# 'listdir' function from 'os'  gives you a list of all
target_file_names_Cri = os.listdir(DATA_DIR_CRI)

critics = list()
for files in target_file_names_Cri:
    target_file_path = os.path.join(DATA_DIR_CRI, files)
    with open(target_file_path,'r') as target_file:
        critics.append(json.load(target_file))
        
```

```python
list_key=critics[1].keys()

count = 0            
for i,crt in enumerate(critics):
    try:
        if(list_key != crt.keys()):
            count += 1
            print i,crt
    except AttributeError:
        count=count
        print i,crt

print count
```

```python
# Removing Attribute Error rows which is not a dictionary
# Removing rows which doesn't have all keys
    
list_index = [80,81,1253,4441]

critics = [j for i,j in enumerate(critics) if i not in list_index]
```

```python
list_key=critics[1].keys()

count = 0            
for i,crt in enumerate(critics):
    try:
        if(list_key != crt.keys()):
            count += 1
            print i,crt
    except AttributeError:
        count=count
        print i,crt

print count
```

```python
# Creating Genre list

genre_list = []

for crt in critics:
    try:
        for i in crt['genre']:
            genre_list.append(i)
    except KeyError:
        print "Key Error"
```

```python
len(genre_list)
```

```python
# Finding unique genre

unique_genre = set(genre_list)
```

```python
len(unique_genre)
```

```python
# Creating dummy variables for each reviews ofr user as well as critic
# Also, addded percentage columns

for i,crt in enumerate(critics):
    try:
        crt['critic_positive'] = crt['num_critic_reviews'][0]
        crt['critic_netural'] = crt['num_critic_reviews'][1]
        crt['critic_negative'] = crt['num_critic_reviews'][2]
        crt['critic_total'] = crt['num_critic_reviews'][3]
        
        crt['user_positive'] = crt['num_user_reviews'][0]
        crt['user_netural'] = crt['num_user_reviews'][1]
        crt['user_negative'] = crt['num_user_reviews'][2]
        crt['user_total'] = crt['num_user_reviews'][3]

        if crt['num_critic_reviews'][3] !=0:
            crt['critic_positive_pct'] = float(crt['num_critic_reviews'][0])/crt['num_critic_reviews'][3]
            crt['critic_netural_pct'] = float(crt['num_critic_reviews'][1])/crt['num_critic_reviews'][3]
            crt['critic_negative_pct'] = float(crt['num_critic_reviews'][2])/crt['num_critic_reviews'][3]
        else:
            crt['critic_positive_pct'] = 0.0
            crt['critic_netural_pct'] = 0.0
            crt['critic_negative_pct'] = 0.0
            
        if crt['num_user_reviews'][3] !=0:
            crt['user_positive_pct'] = float(crt['num_user_reviews'][0])/crt['num_user_reviews'][3]
            crt['user_netural_pct'] = float(crt['num_user_reviews'][1])/crt['num_user_reviews'][3]
            crt['user_negative_pct'] = float(crt['num_user_reviews'][2])/crt['num_user_reviews'][3]
        else:
            crt['user_positive_pct'] = 0.0
            crt['user_netural_pct'] = 0.0
            crt['user_negative_pct'] = 0.0    
    
    except KeyError:
        print i,"Key Error"
```

```python
# Adding columns for each genre and making it 1 if the movie belongs to that genre

for i,crt in enumerate(critics):
    try:
        for gr in unique_genre:
            if gr in crt['genre']:
                crt[gr]=1
            else:
                crt[gr]=0
    except KeyError:
        print "Key Error"
            
```

```python
critics_df=pd.DataFrame(critics)
```

```python
# Creating cleaned title with only character and number

movies_df['cleaned_title'] = movies_df['title'].str.lower()
movies_df['cleaned_title'] = movies_df['cleaned_title'].str.replace("\W",'')

critics_df['cleaned_title'] = critics_df['title'].str.lower()
critics_df['cleaned_title'] = critics_df['cleaned_title'].str.replace("\W",'')
```

```python
# Joining both the data on cleaned title

movie_critics_df = pd.merge(movies_df, critics_df, on='cleaned_title', how='inner')
```

```python
N = 8
top_studio = movie_critics_df.studio.value_counts().index[:N]
top_studio_movies = movie_critics_df[movie_critics_df['studio'].isin(top_studio)]

print '%i movies by top %i studios: %s.' % (len(top_studio_movies),
                                              N,
                                              ', '.join(top_studio))

top_studio_movies.head()
```

```python
studio_dummies = pd.get_dummies(top_studio_movies['studio'])
```

```python
movie_critics_df1 = pd.merge(movie_critics_df, studio_dummies, 
                          left_index=True, 
                          right_index=True,
                          how='left')
```

```python
# Impute missing values to 0 for movies not from top 8 studios

col_list = ['Buena Vista Pictures','Columbia Pictures', 'IFC Films',
           'Paramount Pictures','Sony Pictures Classics',
           'Twentieth Century Fox Film Corporation','Universal Pictures',
           'Warner Bros. Pictures']

movie_critics_df1[col_list] = movie_critics_df1[col_list].fillna(0)
```

```python
N = 8
top_director = movie_critics_df.director_y.value_counts().index[:N]
top_director_movies = movie_critics_df[movie_critics_df['director_y'].isin(top_director)]

director_dummies = pd.get_dummies(top_director_movies['director_y'])

movie_critics_df2 = pd.merge(movie_critics_df1, director_dummies, 
                          left_index=True, 
                          right_index=True,
                          how='left')

# Impute missing values to 0 for movies not from top 8 directors

dir_col_list = ['Clint Eastwood','Ethan Coen','Joel Schumacher',
           'Ridley Scott','Steven Soderbergh',
           'Steven Spielberg','Tim Burton',
           'Woody Allen']

movie_critics_df2[dir_col_list] = movie_critics_df2[dir_col_list].fillna(0)
```

```python
movie_critics_df2.shape
```

```python
#'opening_per_theater','opening_weekend_take'
"""
feature = ['Action','Adult','Adventure',
           'Animation','Biography','Comedy','Crime','Documentary','Drama','Family',
           'Fantasy','Film-Noir','History','Horror','Music','Musical','Mystery','News',
           'Romance','Sci-Fi','Short','Sport','Thriller','War','Western','runtime_minutes',
           'Buena Vista Pictures','Columbia Pictures','IFC Films','Paramount Pictures',
           'Sony Pictures Classics','Twentieth Century Fox Film Corporation',
           'Universal Pictures','Warner Bros. Pictures','Clint Eastwood','Ethan Coen',
           'Joel Schumacher','Ridley Scott','Steven Soderbergh','Steven Spielberg',
           'Tim Burton','Woody Allen']
"""
feature = ['Action','Comedy','Crime','Drama','Romance','Thriller',
           'Buena Vista Pictures','Columbia Pictures','IFC Films','Paramount Pictures',
           'Sony Pictures Classics','Twentieth Century Fox Film Corporation',
           'Universal Pictures','Warner Bros. Pictures','Clint Eastwood','Ethan Coen',
           'Joel Schumacher','Ridley Scott','Steven Soderbergh','Steven Spielberg',
           'Tim Burton','Woody Allen']

feature_response = feature + ['domestic_gross']
```

```python
cleaned_df = movie_critics_df2[feature_response].dropna()

# Check how many movies are dropped because of missing value
movie_critics_df2.shape[0] - cleaned_df.shape[0]
```

```python
X = cleaned_df[feature]
Y = cleaned_df['domestic_gross']

from sklearn.cross_validation import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

print len(X_train), len(X_test)
```

```python
from sklearn.linear_model import ElasticNetCV

model = ElasticNetCV(l1_ratio=[.1, .5, .7, .9, .95, .99, 1],cv=5,verbose=1)

fitted_model = model.fit(X_train,Y_train)
```

```python
fitted_coef = list(fitted_model.coef_)
```

```python
feature_coef_dict = dict()
for index, value in enumerate(fitted_coef):
    feature_coef_dict[feature[index]] = value

feature_coef_dict
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=50,max_depth=5)
fitted_rf_model = rf_model.fit(X_train, Y_train)
```

```python
feature_importance = fitted_rf_model.feature_importances_
feature_import_dict = dict()
for index, value in enumerate(feature_importance):
    feature_import_dict[feature[index]] = value

feature_import_dict
```
